### a) The file "Purchase_Data.csv" contains purchase data at a Chinese supermarket for various products.  Read this csv into a dataframe called df_data and use the head() method to check that you have read in the data correctly. Use dtypes to view column datatypes.  <br>  (Mark: 1)

### b) Slice the first two columns and first two rows using .loc and store the result in a variable called result_1. <br>(Mark: 1)

```python
df_data.shape[0]
```

```python
num_rows = df_data.shape[0]

df_data.loc[num_rows-1:, 'price']
```

### e) Compute the average and max price charged.  Store the results in variables called avg_price and max_price <br> (Mark: 1) 

```python
avg_price = round(df_data['price'].mean(),2)
print("average price charged = $", avg_price)

max_price = df_data['price'].max()
print("max price charged = $", max_price)
```

### f) Create a column called "total", which is the product of the price and num_item columns. <br> (Mark: 1)

```python
df_data['total'] = df_data.num_item * df_data.price

df_data.head(20)
```

### g) Sort the data by 'total'. What is the average price of the three most expensive purchases?  <br> (Mark: 2)

```python
df_data.sort_values('total', ascending=False)
```

```python
df_data.total.nlargest(3).mean()
```

### h) First, covert the column invoice time to a datetime column. Print out the new column types to make sure this was done correctly. Next, let's create columns called "DOW" and "Month", which store the corresponding weekday and month of the invoice respectively.  Print the head of the dataframe to make sure this was done correctly. <br> (Mark: 2)

```python
df_data['Time'] = pd.to_datetime(df_data.invoice_time)
df_data.head()
```

```python
df_data['DOW'] = df_data.Time.dt.day_name()

df_data['Month'] = df_data.Time.dt.month

df_data.head()
```

### i) How many different product_id are there? <br> (Mark: 1)

```python
df_data.product_id.nunique()
```

### j) Which store is the highest total? <br> (Mark: 1)

```python
df_data.groupby(['store_id'])['total'].sum().nlargest(1)
```
